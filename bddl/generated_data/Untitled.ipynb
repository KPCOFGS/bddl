### Remove cookability from melted versions

```python
synsets_old_df = pd.read_csv("synsets_develop.csv")
```

```python
cookables_diff_synsets = synsets_old_df[(synsets_old_df["meltable"] == 1) & (synsets_old_df["cookable"] == 1)]
```

```python
not_cookables = set(cookables_diff_synsets["synset"])
not_cookables_lemmas = set([syn.split("__")[-1].split(".n.")[0] for syn in not_cookables])
not_cookables
```

```python
all_synsets = set(synsets_old_df["synset"])
not_cookables_cooked_deriv_syns = set([syn for syn in all_synsets if ("cooked__" in syn) and any(lemma + ".n." in syn for lemma in not_cookables_lemmas)])
not_cookables_cooked_deriv_syns
```

```python
synsets_df = pd.read_csv("synsets.csv")
cookable_syns = set(synsets_df[synsets_df["cookable"] == 1]["synset"])
synsets_old_df = pd.read_csv("synsets_develop.csv")
old_cookable_syns = set(synsets_old_df[synsets_old_df["cookable"] == 1]["synset"])

deleted_cookables = old_cookable_syns.difference(cookable_syns)
deleted_cookables_lemmas = set([syn.split("__")[-1].split(".n.")[0] for syn in deleted_cookables])
```

```python
invalid_syns = set()
for syn in all_synsets:
    if "cooked__" not in syn:
        continue 
    for lemma in deleted_cookables_lemmas:
        if (lemma + ".n.") in syn:
            print(syn)
            invalid_syns.add(syn)
invalid_syns
```

```python
synsets_df = pd.read_csv("synsets.csv")
synsets_df = synsets_df[~synsets_df["synset"].isin(invalid_syns)]
synsets_df
```

```python
synsets_df.to_csv("synsets.csv", index=False)
```
